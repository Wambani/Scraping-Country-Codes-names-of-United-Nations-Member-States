```python
import pandas as pd
import numpy as np
from urllib.request import urlopen
from bs4 import BeautifulSoup
```

```python
url = ('http://www.fao.org/countryprofiles/iso3list/en/')
html = urlopen(url)
```

```python
soup = BeautifulSoup(html, 'lxml')
type(soup)
```

```python
# Getting the title of the page
title = soup.title
print(title)
```

```python
content = soup.find(id="c460959")
```
